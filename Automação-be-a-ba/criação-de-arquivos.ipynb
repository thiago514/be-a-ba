### Formatação

```python
import sys
from faker import Faker
import random

fake = Faker('pt_BR')
palavras = []

with open("palavras.txt", "r") as f:
    dados = f.read()
    palavras = dados.split(",")

def formato_celula(formato: str):
    if formato == 'datetime':
        return 'yyyy-mm-dd'
    elif formato == 'bool':
        return 'BOOLEAN'
    elif formato == 'int':
        return '0'
    elif formato == 'float':
        return '0.00'
    else:
        return '@'
    
def valor_celula(formato: str):
    
    if formato == 'datetime':
        return fake.date(pattern="%Y-%m-%d", end_datetime=None)
    elif formato == 'bool':
        return fake.boolean()
    elif formato == 'int':
        return random.randint(0, pow(2,32))
    elif formato == 'float':
        return random.uniform(0, pow(2, 16))
    else:
        return fake.word(ext_word_list=palavras)
```

### Função para gerar arquivo XSLX

```python
import random
import pandas as pd
import io
import xlsxwriter

def gerar_template_populado_xlsx(template):
    print("gerando arquivo")
    print(template)
    buffer = io.BytesIO()
    buffer.name = template["nome"] + '.' + template["extencao_do_arquivo"]
    workbook = xlsxwriter.Workbook(buffer)
    for tabela in template["tabelas"]:
        print("tabela " + tabela["nome"])
        worksheet = workbook.add_worksheet(tabela["nome"])
        num_linhas = random.randint(1000, 10000)
        for i, campo in enumerate(tabela["campos"]):
            print("campo " + campo["nome"])
            print("tipo " + campo["tipo"])
            worksheet.write(0, i, campo["nome"])
            for n in range(1, num_linhas):
                formato = workbook.add_format()
                formato.set_num_format(formato_celula(campo["tipo"]))
                worksheet.write(n, i, valor_celula(campo["tipo"]), formato)
    workbook.close()
    return buffer



```

### Gerando arquivo CSV

```python
import random
import pandas as pd
import io
def gerar_template_populado_csv(template):
    print("gerando arquivo")
    print(template)
    buffer = io.BytesIO()
    buffer.name = template["nome"] + '.' + template["extencao_do_arquivo"]
    df = pd.DataFrame()
    num_linhas = 10000
    for i, campo in enumerate(template["tabelas"][0]["campos"]):
        df[campo["nome"]] = gerando_lista_valores(campo["tipo"], num_linhas)
    df.to_csv(buffer, index=False)
    return buffer

def gerando_lista_valores(tipo, tamanho):
    lista = []
    for i in range(tamanho):
        lista.append(valor_celula(tipo))
    return lista



```

### Gerando Arquivo para todos os templates

```python
arquivos = os.listdir("./templates/")
for arquivo in arquivos:
    template = json.load(open("templates/" + arquivo, "r"))
    buffer = None
    if template["extencao_do_arquivo"] == "CSV":
        buffer = gerar_template_populado_csv(template)
    else:
        buffer = gerar_template_populado_xlsx(template)
        
    with open("arquivos/"+ buffer.name.strip(), "wb") as file:
        file.write(buffer.getbuffer())
```
